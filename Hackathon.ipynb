### Extracting the matrix of features

```python
# X = heart.drop("num")
# y = heart["num"].value
# heart.columns[0:-1]
X = heart[heart.columns[0:-1]].values
y = heart["num"].values
```

```python
print(X.shape, y.shape)
```

### Performing feature scaling on X

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
X = sc_X.fit_transform(X)
```

### Doing the train test split

```python
y = np.reshape(y, (-1,1))

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

```python
y.shape
```

### The classification model

```python
from keras.layers import Dense, Dropout
from keras.models import Sequential
```

```python
#getting the number of feature columns
n_cols = X.shape[1]



# model building start
model = Sequential()

# you only specify the input shape of the 1st perceptron for it to know the number of features to expect 
# subsequent perceptrons calculate their input_shape based on the number of units of the previous perceptron
model.add(Dense(100, activation="relu", input_shape=(n_cols, )))
model.add(Dropout(0.2))
model.add(Dense(100, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(100, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

### compiling the model

```python
# investigate the different optimers and loss functions and their meanings
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

### Fitting the model

```python
model.fit(X_train, y_train, batch_size=32, epochs=100)
```

```python
#using the model
predictions = model.predict_classes(X_test)
```

### Assessing the model

```python
#Analyze these codes well to get meaning
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python

print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
